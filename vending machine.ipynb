```python
new_data = data[['개방서비스명', '소재지전화', '소재지전체주소', '도로명전체주소', '사업장명']]
```

```python
new_data = new_data[new_data['사업장명'].str.contains('복지')]
```

```python
new_data.reset_index(drop=True, inplace=True)
```

```python
new_data
```

```python
new_data['소재지전화'] = new_data['소재지전화'].fillna('-')
```

```python
new_data['도로명전체주소'] = new_data['도로명전체주소'].fillna('-')
```

```python
new_data
```

```python
# import 라이브러리
from geopy.geocoders import Nominatim
from tqdm import tqdm

#예제 데이터 : df_shake
#컬럼 정보 : name, branch, addr

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo_local = Nominatim(user_agent='South Korea')  #지역설정
        location = geo_local.geocode(address)
        geo = [location.latitude, location.longitude]
        return geo

    except:
        return [0,0]

# 실행
for idx, addr in enumerate(tqdm(new_data.도로명전체주소)):
    new_data.loc[idx,'좌표정보(x)'] = geocoding(addr)[0]
    new_data.loc[idx,'좌표정보(y)'] = geocoding(addr)[1]
```
