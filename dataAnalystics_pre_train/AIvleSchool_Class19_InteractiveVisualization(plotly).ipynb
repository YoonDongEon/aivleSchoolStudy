## <span style="background-color:#F1F8FF">add_trace</span> :  캔들스틱 차트 구현

```python
fig.add_trace(
    go.Candlestick(     # go.Candlestick 그래프
        x=df.index,     # x축은 날짜로 표기
        open=df.Open,   # 시초가
        high=df.High,   # 고가
        low=df.Low,     # 저가
        close=df.Close, # 종가
        increasing_line_color='red',    # 주가 상승은 빨간색
        decreasing_line_color='blue',   # 주가 하락은 파란색
    )
)
```

## <span style="background-color:#F1F8FF">update_layout</span> :  그래프 레이아웃 설정

```python
print(fig)
```

```python
# Figure 객체를 출력해보면 'Figure({'data': [], 'layout': {'template': '...'}})' 이런 구조의 속성 값이 나온 것을 확인할 수 있음
# Figure 객체를 최초 생성시 data에 대한 값을 주는 방법으로 캔들스틱을 구현 가능
# 'plotly'의 'layout'은 다양하기에 보다 간편하게 'updata_layout' 메서드로 업데이트 진행

fig = go.Figure(
    data = [go.Candlestick(     # go.Candlestick 그래프
        x=df.index,     # x축은 날짜로 표기
        open=df.Open,   # 시초가
        high=df.High,   # 고가
        low=df.Low,     # 저가
        close=df.Close, # 종가
        increasing_line_color='red',    # 주가 상승은 빨간색
        decreasing_line_color='blue',   # 주가 하락은 파란색
    )]
)

fig.show()
```

```python
# graph_objects의 그래프에 타이틀과 축 레이블을 표기하는 방법 -> update_layout 사용
fig.update_layout(title='Stock Price of XXX(OHLC, $)', yaxis_title='$',)

fig.layout.height = 600
fig.show()
```

## 이동 평균성 생성(Moving Average)

```python
# 종가를 이용한 MA
ma_3 = df.Close.rolling(3).mean() # 3개씩 밀어서 평균
ma_10 = df.Close.rolling(10).mean() # 10개씩 밀어서 평균
ma_30 = df.Close.rolling(30).mean() # 30개씩 밀어서 평균
```

```python
df['MA3'] = ma_3
df['MA10'] = ma_10
df['MA30'] = ma_30
```

```python
display(df)
```

```python
# moving average (이동 평균을 라인 그래프로 표현)

ma3 = go.Scatter(
    x = df.index, # 날짜
    y = df.MA3, # 이동 평균
    line = dict(color='black', width=0.5), # 검정색, 0.5 굵기
    name = 'MA(3)'
)

ma10 = go.Scatter(
    x = df.index,
    y = df.MA10,
    line = dict(color='red', width=0.5),
    name = 'MA(10)'
)

ma30 = go.Scatter(
    x = df.index,
    y = df.MA30,
    line = dict(color='green', width=0.5),
    name = 'MA(30)'
)
```

```python
# 기존 fig객체에 ma3 추기
fig.add_traces(ma3)
fig.show()
```

확대 시 이동 평균선이 생긴 것을 확인 가능

```python
# ma10과 ma30 추가
fig.add_traces(ma10)
fig.add_traces(ma30)
fig.show()
```

### Bollinger Bands 추가
볼린저 밴드는 주가의 변동이 표준정규분포를 따른다고 가정하고, 주가의 추세를 따라 위아래로 폭이 같이 움직이는 밴드를 만들어 주가를 그 밴드 기준선으로 판단하고자 만들어진 지표이다.
\※ 만드는 원리
 1. 기준선을 구한다. 여기서는 <span style="background-color:#F1F8FF">ma30</span> (이동평균 30일)으로 설정
 2. 기준선에 대응하는 표준편차를 구함 -> <span style="background-color:#F1F8FF">df.Close.rolling(30).std</span>
 3. 상한선 = 기준선 + (2 * 표준편차)
 4. 하한선 = 기준선 - (2 * 표준편차)

```python
# 표준편차 구하기
std_30 = df.Close.rolling(30).std()
```

```python
df['STD30'] = std_30
```

```python
# Bollinger Bands 그리기

# 상한선
upper_bound = go.Scatter(
    x = df.index,
    y = df.MA30 + (df.STD30 * 2),
    line = dict(dash='dash', color='gray', width=1),
    name = 'upper bounds'
)

# 하한선
lower_bound = go.Scatter(
    x = df.index,
    y = df.MA30 - (df.STD30 * 2),
    line = dict(dash='dash', color='gray', width=1),
    fill = 'tonexty',   # 상한선과 하한선 사이 면적 채우기
    opacity = 0.5,  # 투명도 조절
    name = 'lower bounds'
)
```

```python
# 캔들스틱에 상한선, 하한선 추기
fig.add_traces([upper_bound, lower_bound])
```

# 바 그래프: 상한선 추가

```python
# 거래량을 나타내는 trading_volume의 바 플롯
trading_volume = go.Bar(x=df.index, y=df.Volume)
```

```python
import plotly.subplots as ms # 거래량 바그래프 추가를 위해 plotly의 subplots 패키지 로딩
```

```python
# (2 x 1)의 Axes를 갖고 있는 figure 생성

fig = ms.make_subplots(
    rows=2, # 2행
    cols=1, # 1열
    shared_xaxes=True, # x축 값 공유
    # vertical_spacing=0.02
    subplot_titles=('Stock Prices of XXX', 'Trading Volumes')   # 두 개의 플롯에 타이틀 생성(첫번째 플롯, 두번째 플롯)
)
```

```python
fig
```

```python
# 캔들스틱 그래프 추가 (add_trace를 사용함, add_trace's' 아님!)
fig.add_trace(
    # 캔들스틱 추가
    go.Candlestick(
        x = df.index, # 날짜
        open = df.Open, # 시초가
        high = df.High, # 고가
        low = df.Low, # 저가
        close = df.Close, # 종가
        increasing_line_color = 'red', # 상승은 빨간색
        decreasing_line_color = 'blue', # 하락은 파란색
    ),
    # 몇번째 행과 열에 그래프를 그릴건지 명시
    row = 1, # 1행
    col = 1 # 1열
)

# figure 높이 설정
fig.layout.height = 600
3fig.show()
```

```python
# 이동평균선 추가 (1행 1열 그래프에 추기)
fig.add_trace(ma3, row=1, col=1)
fig.add_trace(ma10, row=1, col=1)
fig.add_trace(ma30, row=1, col=1)
fig.show()
```

```python
# 볼린저 라인 차게
fig.add_trace(upper_bound, row=1, col=1) # 상한선
fig.add_trace(lower_bound, row=1, col=1) # 하한선
```

```python
# 거래량 추가 2행 1열 - 즉 두번째 그래프에 추가
fig.add_trace(trading_volume, row=2, col=1)
fig.show()
```

```python
# figure 레이아웃 설정
fig.update_layout(
    yaxis1_title = 'Stock Price (KRW)', # 첫번째 캔들스틱에 대한 y축 레이발
    yaxis2_title = 'Trading Volumes', # 두번째 트레이닝 바 차트에 대한 y축 레이블
    xaxis2_title = 'Periods', # 두번째 그래프 x축 레이블
    xaxis1_rangeslider_visible = False, # 첫번째 그래프 레인지 슬라이더 숨김
    xaxis2_rangeslider_visible = True, # 두번째 그래프 레인지 슬라이더 활성화
)

fig.show()
```
