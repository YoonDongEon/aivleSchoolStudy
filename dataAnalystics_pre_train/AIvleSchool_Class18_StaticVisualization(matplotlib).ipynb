# 정적 시각화 실습 - <span style="background-color:#F1F8FF">matplotlib</span> 활용
### <span style="background-color:#F1F8FF">pyplot</span>의 <span style="background-color:#F1F8FF">plot</span>

```python
import matplotlib.pyplot as plt # pyplot 모듈
import numpy as np

# 브라우저 내부(inline)에 바로 그려지도록 해주는 코드
%matplotlib inline
```

```python
# 정규분포에서 난수를 추출
np.random.seed(0) # 동일한 난수가 나오게끔 시드 설정

x = np.arange(50) # 0부터 49까지 1씩 증가하는 정수 생성
y = np.random.randn(50) # 정규분포 난수 생성
print(y)
```

### 기본 그래프 그리기
 1. figure를 만들고 --> <span style="background-color:#F1F8FF">plt.figure()</span>
 2. figure내 그래프를 그리고 --> <span style="background-color:#F1F8FF">plt.plot(x, y)</span>
 3. 그래프를 출력 --> <span style="background-color:#F1F8FF">plt.show()</span>

```python
# 기본적인 그래프 그리는 방법 plot(x, y)
plt.figure()
plt.plot(x, y)
plt.show()
```

### <span style="background-color:#F1F8FF">figure</span> 크기 설정 & 축 제목 표기

```python
# 타이틀과 x축, y축 레이블 생성

plt.figure(figsize=(10, 5)) # figsize 인수를 사용하여 figure의 크기 조정 (가로길이, 세로길이)
plt.plot(x, y)
plt.xlabel('Period (t)') # xlabel을 사용하면 x축 레이블을 표기할 수 있음
plt.ylabel('Percent change') # ylabel을 사용하면 y축 레이블을 표기할 수 있음
plt.title('Change of Interest Rate over Last 50 Days') # title을 사용하면 그래프 제목 설정 가능
# plt.savefig('./data/interest_rate.png)
plt.show() # figure를 출력
```

### 그래프에 데이터 추가

```python
np.random.seed(2) # 고정된 난수 설정

z = np.random.randn(50) # 정규분포에서 50개 추출

print(z)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Korea') # label을 이용하여 각 선이 어떤 데이터를 나타내는지 확인
plt.plot(x, z, label='United States')
plt.xlabel('Period (t)')
plt.ylabel('Percent change')
plt.title('Change of Interest Rate over Last 50 Days')
plt.legend() # legend를 이용하면 범례 표기 가능
plt.tight_layout() # 그림을 사이즈에 맞게 변경하여 출력
# plt.savefig('./data/interest_rate.png)
plt.show()
```

## 여러 그래프 그리기 1: <span style="background-color:#F1F8FF">subplot</span>
<span style="background-color:#F1F8FF">matplotlib</span>의 <span style="background-color:#F1F8FF">pyplot</span>이 제공하는 <span style="background-color:#F1F8FF">subplot</span> 메서드는 여러개의 그래프를 하나의 그림에 나타냄

<span style="background-color:#F1F8FF">subplot</span>은 그림의 위치와 순서에 관련한 3개 값을 설정해야 함. -> plt.subplot(행, 열, 인덱스)

```python
a = np.linspace(0.0, 3.0, 30) # 0과 3사이에 있는 실수 30개를 균등한 간격으로 생성
print(a)
```

```python
sin = np.sin(2 * np.pi * a)
print(sin)

decaying_sin = np.sin(2 * np.pi * a) * np.exp(-a)
print(decaying_sin)
```

```python
plt.subplot(2, 1, 1) # 2개의 행과 1개의 열을 사용하여 첫번째(위) 칸에 그리겠다고 명시
plt.plot(a, decaying_sin, '-g.')    # 라인 : -, 초록색 : g, 도트 : . 으로 설정
plt.title('Sine Graph Approaching to Zero')

plt.subplot(2, 1, 2) # 2개의 행과 1개의 열을 사용하여 두번째(밑) 같에 그리겠다고 명시
plt.plot(a, sin, 'o-')  # 도트 : o, 라인 : - 으로 설정
plt.title('Sine Graph')

plt.tight_layout()
# plt.savefig('sine_graphs.png') # 그래프를 저장하는 커맨드
plt.show()
```

### <span style="background-color:#F1F8FF">sharex</span>: x축 공유

```python
plt.subplot(2, 1, 1) # 2개의 행과 1개의 열을 사용하여 첫번째(위) 칸에 그리겠다고 명시
plt.plot(a, decaying_sin, '-g.', label='decaying_sin')
plt.xticks(visible=False) # visible=False 설정
plt.title('Sine Graph Approaching to Zero')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(a, sin, 'o-', label='sine')  # 도트 : o, 라인 : -
plt.title('Sine Graph')
plt.xlabel('Num Cycle')
plt.legend()

plt.suptitle('Two Different Sine Graph', y=1.05, fontsize=20) # subtitle을 이용하여 제목 설정

plt.show()
```

## 여러 그래프 그리기 2: <span style="background-color:#F1F8FF">subplots</span>

```python
fig, ax = plt.subplots(2, 1) # figure과 ax를 같이 설정해줘야 함
# ax
ax[0].plot(a, decaying_sin, 'g-') # ax[행번호, 열번호], 지금은 2개의 그래프이기에 인덱스 하나로 표현 가능
ax[1].plot(a, sin, 'b-.')
plt.show()
```

<span style="background-color:#F1F8FF">subplot</span>과 <span style="background-color:#F1F8FF">subplots</span>의 차이는 <span style="background-color:#F1F8FF">state-based</span>와 <span style="background-color:#F1F8FF">object-oriented</span> 방식의 차이와 같음

### 두 종류의 그래프를 하나에 담기: 바 그래프 & 라인 그래프
아래의 데이터는 임의로 만든 ABC라는 가상의 국가의 1951년부터 1960년까지 집 거래량과 연 평균 집값을 의미
 * <span style="background-color:#F1F8FF">연 평균 집값 (housing_prices)</span>
 * <span style="background-color:#F1F8FF">총 거래량 (num_trading)</span>
 * <span style="background-color:#F1F8FF">기간 (periods)</span>

```python
# 1. 토이 데이터 준비
# ABC 국가의 연간 평균 집값과 총 거래량
periods = np.arange(1951, 1961) # 연도(1951년부터 1960년까지 생성)
housing_prices = np.array([15, 25, 31, 24, 22, 27, 40, 55, 60, 58]) # Unit = $'000
num_trading = np.array([6, 6.6, 7, 4.5, 4, 6, 9.5, 11, 20, 18]) # unit = 백만
```

<span style="background-color:#F1F8FF">plt.subplots</span>를 사용하려면 figure를 나타낼 <span style="background-color:#F1F8FF">fig</span> 변수와 Axes 객체를 나타낼 <span style="background-color:#F1F8FF">ax</span> 변수가 필요

Axes 객체의 함수 중 하나인 <span style="background-color:#F1F8FF">twinx</span> 메서드는 특정 <span style="background-color:#F1F8FF">ax</span>와 x축을 공유하는 새로운 Axes를 만들때 사용함 (ex. <span style="background-color:#F1F8FF">ax2 = ax1.twinx()</span>)

```python
fig, ax1 = plt.subplots(figsize=(9, 5)) # figsize 인수를 주어 figure의 크기를 정의

# ax1
# 집값 그래프는 라인으로 표기 plot 이용
ax1.plot(periods, housing_prices, 'o-', linewidth=1, alpha=0.8, label='housing price') # 'o-': 데이터마다 o로 마킹된 라인
ax1.set_xlabel('Year', fontsize=13) # ax1에 대해 x축 (y축) 레이블을 설정하려면 set_xlabel (set_ylabel)을 사용
ax1.set_ylabel('Price ($000)', fontsize=13) # 단위는 천달러부터 e.g. 6 : 6천 달러, 50 : 5만 달러, etc.
ax1.set_ylim(0, 70) # ax1의 y축 범위 설정. 여기서는 0부터 70까지만 표기

# 총 거래량 그래프는 막대 그래프(bar plot)로 표기
ax2 = ax1.twinx()   # twinx는 x축은 공유하지만 y축은 공유하지 않게 함
ax2.bar(periods, num_trading, color='coral', alpha=0.3, label='#houses traded') # 막대 그래프 생성
ax2.set_ylabel('#house traded (millons)', fontsize=13)
ax2.set_ylim(0, 22)

ax1.legend(loc='upper left') # legend에 loc 인수를 설정하면 범례의 위치를 설정 가능
ax2.legend(loc='upper right')

plt.title('Housing Prices and Demands of ABC Country between 1951 and 1960', fontsize=16)
plt.tight_layout()
# plt.savefig('two_in_one.png') # 해당 그래프를 png 파일로 저장

plt.show()
```
